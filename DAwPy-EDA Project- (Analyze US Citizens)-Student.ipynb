 # What country has the highest percentage of people that earn >50K?
 ---


```python
country_salary_df = df.groupby('native-country')[['salary']].count()
country_salary_df
```

```python
df.groupby(['native-country', 'salary'])[['salary']].count()
```

```python
df.loc[df["salary"] == ">50K", "native-country"].value_counts()
```

```python
country_salary_df_riches = df.groupby(['native-country', 'salary'])[['salary']].count().iloc[df.groupby(['native-country', 'salary'])[['salary']].count().index.isin(['>50K'], level=1)].reset_index(drop=True, level=1)
country_salary_df_riches
```

```python
rich_per_country = country_salary_df_riches/country_salary_df
rich_per_country
```

```python
rich_per_country[rich_per_country['salary'] == rich_per_country['salary'].max()]
```

```python
#use iloc here
rich_per_country.iloc[list(rich_per_country['salary'] == rich_per_country['salary'].max())]
```

```python
plt.figure(figsize=(14,7))
df.loc[df["salary"] == ">50K", "native-country"].value_counts().plot(kind="bar");
```

```python
plt.figure(figsize=(12,6))
df.loc[((df["native-country"] == "Iran")), "salary"].value_counts().plot.bar();
```
